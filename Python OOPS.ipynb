### How to manipulate Data in class

```python
class Compid:
    
    def __init__(self):
        self.name = 'Shavej'
        self.age  =  21
        
    def update(self):
        self.age = 22
        
    def compare(self,other):
        if self.age == other.age:
            return True
        else:
            return False
        

com1 = Compid()
com1.age = 30
com2 = Compid()

if com1.compare(com2):
    print("There are same")
else:
    print("There are Different")


print(id(com2))
print(id(com1))

com1.name = 'Shaikh'

print(com1.age);
com1.update()


print(com1.name);
print(com1.age);
print(com2.name);
```

## Class Method Static Method

```python
class Student:
    
    school = "Shavej"
    
    def __init__(self,m1,m2,m3):
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3

    def avg(self):
        return(self.m1+self.m2+self.m3)/3
    
    @classmethod
    def getSchool(cls):
        return cls.school
    
    @staticmethod
    def info():
        print("This is Student class ..,")
    
    
s1 = Student(34,55,25)
s2 = Student(89,32,12)

print(s1.avg())
print(s2.avg())

print(s1.m1)# Accessor

print(Student.getSchool())

Student.info()
```

# Inheritance 

### Single level and Multi level Inheritance


```python
class A:
    def feature1(self):
        print("Feature 1 working")
    
    def feature2(self):
        print("Feature 2 working")

class B(A):
    def feature3(self):
        print("Feature 3 working")
    def feature4(self):
        print("Feature 4 working")
        
class C(B):
    def feature5(self):
        print("Feature 5 working")
        
      
a1 = A()
a1.feature1()
a1.feature2()

b1 = B()
b1.feature1()

c1 = C()
c1.feature1()
```

### Multiple Level Inheritance

```python
# Multiple Level Inheritance
class A:
    def feature1(self):
        print("Feature 1 working")
    
    def feature2(self):
        print("Feature 2 working")

class B:
    def feature3(self):
        print("Feature 3 working")
    def feature4(self):
        print("Feature 4 working")
        
class C(A,B):
    def feature5(self):
        print("Feature 5 working")

a1 = A()
a1.feature1()
a1.feature2()

b1 = B()
# b1.feature1()

c1 = C()
c1.feature1()
c1.feature3()
```

## Constructor in Inheritance
## Method Resolution Order(MRO)

```python
class A:
    
    def __init__(self):
        print("Print A")
    def feature1(self):
        print("Feature 1-A working")
    
    def feature2(self):
        print("Feature 2 working")

class B:
    
    def __init__(self):
        #Call A constructor
#         super().__init__()
        print("Print B")
        
    def feature1(self):
        print("Feature 3-B working")
    def feature4(self):
        print("Feature 4 working")
        
# MRO
class C(A,B):
    def __init__(self):
        super().__init__()
        print('Print C')
        
    def feat(self):
        super().feature2()
      

# a1 = A()
# a1.feature1()
# a1.feature2()

# b1 = C()
# b1.feature1()

c1 = C()
c1.feature1()
c1.feat()
# c1.feature3()
```

### 2. Operator Overloading
--> When Addition or Subtraction method(function) is call then it will class __add__ and __sub__ __mul__ etc

```python
class Student:
    
    def __init__(self,m1,m2):
        self.m1=m1
        self.m2=m2
    
    def __add__(self,other):
        m1 =self.m1 + other.m1
        m2 =self.m2 + other.m2
        s3 = Student(m1,m2)
        return s3
    
    def __gt__(self,other):
        r1 = self.m1 + self.m2
        r2 = self.m1+ self.m2
        
        if r1> r2:
            return True
        else:
            return False
        
    def __str__(self):
        return '{} {}'.format(self.m1,self.m2)
    
s1 = Student(58,69)
s2 = Student(60,65)

s3 = s1 + s2

print(s3.m2)
print(s1.__str__())

if s1 > s2:
    print("s1 win")
else:
    print("s2 win")
    
a = 9
print(id(a))
print(a.__int__())
```

### 3. Method Overloading

```python
class Student:
    
    def __init__(self,m1,m2):
        self.m1=m1
        self.m2=m2
    
    def sum(self,a=None,b=None,c=None):
        s=0
        if a!=None  and b!=None and c!=None:
            s = a+b+c
        elif a!=None  and b!=None:
            s = a+b
        else:
            s = a
        return s
    
    
s1 = Student(58,69)
s2 = Student(60,65)

print(s1.sum(5,9))
print(s1.sum(5,9,10))
print(s1.sum(5))
```
